# Multivariate Gaussian with full covariance

- TensorFlow를 사용하여 다변량 가우시안 분포를 지정하는 방법을 배웁니다.  

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

import tensorflow as tf
import tensorflow_probability as tfp
tfd = tfp.distributions
```

## tfd.Normal

- loc 및 scale 매개변수가 있는 정규 분포입니다.

```python
# 2 개의 scalar 값 정규분포 batch 정의

dist = tfd.Normal(loc=[1, 2.], scale=[11, 22.])

dist.mean(), dist.stddev()
```

- 위에서 구한 dist 분포로 부터 data 를 sampling

```python
# 2d 구형 Gaussian 의 approximate density contour 를 plot
N = 1000
x = dist.sample(N)
x1 = x[:, 0]
x2 = x[:, 1]
sns.jointplot(x=x1, y=x2, kind = 'kde', space = 0)
```

## tfd.Independent
- 분포의 batch 로 부터의 독립 분포 

- reinterpreted_batch_ndims: event dims 로 간주될 가장 오른쪽 batch dims 값 (integer). 'None' 으로 지정하면 첫 번째 배치 축(batch axis 0)을 제외한 모든 것이 이벤트 차원으로 변환됩니다(`tf.layers.flatten`과 유사).

```python
# 2-batch Normal 로 부터 독립 분포 작성

ind = tfd.Independent(
    distribution=tfd.Normal(loc=[1, 2.], scale=[11, 22.]),
    reinterpreted_batch_ndims=1)

ind.mean(), ind.stddev()
```

위에서 구한 ind 분포로 부터 data sampling

```python
# 2d 구형 Gaussian 의 approximate density contour 를 plot
N = 1000
x = ind.sample(N)
print(x.shape)

x1 = x[:, 0]
x2 = x[:, 1]
sns.jointplot(x=x1, y=x2, kind = 'kde', space = 0)
```

## tfd.MultivariateNormalDiag

- $R^k$ 에 대한 다변수 정규 분포

`tfd.MultivariateNormalDiag` 클래스를 사용하면 대각 공분산 행렬 $\Sigma$를 사용하여 다변량 가우시안을 지정할 수 있습니다.

분산이 각 성분에 대해 동일한 경우, 즉 $\Sigma=\sigma^2I$인 경우, 이것은 구형 또는 등방성(isotropic) 가우시안으로 알려져 있습니다. 이 이름은 확률 밀도 함수의 구형(또는 원형) 등고선에서 비롯된 것입니다. 2 차원 사례에 대한 아래 그림에서 볼 수 있습니다.

```python
# 한개의 다변량 정규분포 정의

mvn = tfd.MultivariateNormalDiag(
    loc=[1., 2.],
    scale_diag=[11., 22.])

mvn.mean(), mvn.stddev()
```

```python
# 확률 밀도/질량 함수
mvn.prob([-1., 0])
```

위에서 작성한 다변량 독립 분포에서 data sampling

```python
# 2d 구형 Gaussian 의 approximate density contour 를 plot
N = 1000
x = mvn.sample(N)
print(x.shape)

x1 = x[:, 0]
x2 = x[:, 1]
sns.jointplot(x=x1, y=x2, kind = 'kde', space = 0)
```

```python
# 2-batch 이변량 정규분포에서 독립 분포를 만듭니다.

ind = tfd.Independent(
    distribution=tfd.MultivariateNormalDiag(
        loc=[[-1., 1], [1, -1]],
        scale_identity_multiplier=[1., 0.5]),
    reinterpreted_batch_ndims=1)
ind
```

위에서 작성한 2 변량 독립 분포에서 data sampling

```python
# 2d 구형 Gaussian 의 approximate density contour 를 plot
N = 1000
x = ind.sample(N)
print(x.shape)

x1 = x[:, 0, 0]
x2 = x[:, 0, 1]
sns.jointplot(x=x1, y=x2, kind = 'kde', space = 0)

x1 = x[:, 1, 0]
x2 = x[:, 1, 1]
sns.jointplot(x=x1, y=x2, kind = 'kde', space = 0)
```

# Full covariance with MultivariateNormalFullTriL

- tfd.MultivariateNormalTriL 분포를 사용하여 TensorFlow에서 전체 공분산 가우시안 분포를 정의할 수 있습니다.

- 수학적으로 다변량 가우시안의 매개 변수는 평균 $\mu$와 공분산 행렬 $\Sigma$입니다. 따라서 tfd.MultivariateNormalTriL 생성자에는 두 개의 인수가 필요합니다.
    - loc : $\mu$ 에 해당하는 float Tensor
    - scale_tril : 하부 삼각 행렬 $L$ ($LL^T=\Sigma$)  
    
- d 차원 random variable 의 경우 lower-triangle matrix $L$은 다음과 같습니다.
<img src="lower-triangle.png" width=200>

- 다음은 비 대각선 공분산(non-diagonal covariance)으로 2 차원 가우시안을 생성하는 예입니다.

```python
# Set the mean and covariance parameters
mu = [0., 0.]  # mean
scale_tril = [[1.,  0.], [0.6, 0.8]]  # covariance

# covariance matrix
sigma = tf.matmul(tf.constant(scale_tril), tf.transpose(tf.constant(scale_tril)))  
print(sigma)
```

```python
# Create the 2D Gaussian with full covariance
nonspherical_2d_gaussian = tfd.MultivariateNormalTriL(loc = mu, scale_tril = scale_tril)
nonspherical_2d_gaussian
```

```python
# Check the Distribution mean
print(nonspherical_2d_gaussian.mean())

# Check the Distribution covariance
print(nonspherical_2d_gaussian.covariance())
```

```python
# Plot its approximate density contours
x = nonspherical_2d_gaussian.sample(N)
x1 = x[:, 0]
x2 = x[:, 1]
sns.jointplot(x=x1, y=x2, kind = 'kde', space = 0, color = 'r')
```

- 보시다시피, 대략적인 밀도 윤곽은 이제 원형이 아니라 타원형입니다. 이는 가우시안의 구성 요소가 상관 관계가 있기 때문입니다.  


- 또한 주변 분포(그림의 측면에 표시됨)는 둘 다 단변량 가우시안 분포입니다.

```python
# Load data.
n = 1000
scale_tril = np.array([[1.6180, 0.],
                                [-2.7183, 3.1416]]).astype(np.float32)
x = tfd.Normal(loc=0, scale=1).sample([n, 2])
eps = tfd.Normal(loc=0, scale=0.01).sample([n, 2])
y = tf.matmul(x, scale_tril) + eps
y.shape, y[:10]
```

```python
# Create model.
d = y.shape[-1]

model = tf.keras.Sequential([
    tf.keras.layers.Dense(tfp.layers.MultivariateNormalTriL.params_size(d)),
    tfp.layers.MultivariateNormalTriL(d),
])
```

```python
# Fit.
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.02),
              loss=lambda y, model: -model.log_prob(y), 
              metrics=[])

batch_size = 100
model.fit(x, y, batch_size=batch_size ,
          epochs=50, steps_per_epoch=n // batch_size,
          verbose=0, shuffle=True)
```

```python
xhat = model(x[0:1])
```
