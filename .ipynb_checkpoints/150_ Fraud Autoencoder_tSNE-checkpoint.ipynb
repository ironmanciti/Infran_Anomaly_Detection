# Autoencoder for Fraud Detection - t-SNE

```python
import numpy as np 
import pandas as pd 
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data/creditcard.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.drop('Time',axis=1)
```

```python
X = df.drop('Class',axis=1).values 
y = df['Class'].values
```

```python
X.shape
```

- feature scaling

```python
X -= X.min(axis=0)
X /= X.max(axis=0)
```

```python
X.mean()
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

- 필요에 따라 다른 활성화 함수를 사용할 수 있음

```python
data_in = Input(shape=(29,))
encoded = Dense(12, activation='tanh')(data_in)  
decoded = Dense(29, activation='sigmoid')(encoded)
autoencoder = Model(data_in,decoded)
```

```python
autoencoder.compile(optimizer='adam', loss='mean_squared_error')
```

```python
autoencoder.fit(X_train, X_train,
                epochs = 20, 
                batch_size = 128, 
                validation_data=(X_test,X_test))
```

```python
X_test.mean()
```

```python
pred = autoencoder.predict(X_test[0:10])
```

- original data 와 reconstruction data 비교

```python
import matplotlib.pyplot as plt
import numpy as np

width = 0.8

prediction   = pred[9]
true_value    = X_test[9]

indices = np.arange(len(prediction))

fig = plt.figure(figsize=(10,7))

plt.bar(indices, prediction, width=width, color='b', label='Predicted Value')

plt.bar([i+0.25*width for i in indices], true_value, 
        width=0.5*width, color='r', alpha=0.5, label='True Value')

plt.xticks(indices+width/2., 
           ['V{}'.format(i) for i in range(len(prediction))] )

plt.legend()

plt.show()
```

- 보시다시피, 우리 모델은 원래 값을 재구성하는 데 훌륭한 역할을합니다. 시각적 검사는 추상적인 숫자보다 더 많은 통찰력을 제공합니다.

# VAE

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

from tensorflow.keras.layers import Input, Dense, Lambda
from tensorflow.keras.models import Model
from tensorflow.keras import backend as K
from tensorflow.keras import metrics
```

```python
batch_size = 100
original_dim = 29
latent_dim = 6
intermediate_dim = 16
epochs = 50
epsilon_std = 1.0
```

```python
x = Input(shape=(original_dim,))
h = Dense(intermediate_dim, activation='relu')(x)
z_mean = Dense(latent_dim)(h)
z_log_var = Dense(latent_dim)(h)
```

```python
def sampling(args):
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim), mean=0.,
                              stddev=epsilon_std)
    return z_mean + K.exp(z_log_var / 2) * epsilon
```

```python
# note that "output_shape" isn't necessary with the TensorFlow backend
z = Lambda(sampling, output_shape=(latent_dim,))([z_mean, z_log_var])
```

```python
# we instantiate these layers separately so as to reuse them later
decoder_h = Dense(intermediate_dim, activation='relu')
h_decoded = decoder_h(z)

decoder_mean = Dense(original_dim)
x_decoded_mean = decoder_mean(h_decoded)
```

```python
# instantiate VAE model
vae = Model(x, x_decoded_mean)
```

```python
# Compute VAE loss
xent_loss = original_dim * metrics.mean_squared_error(x, x_decoded_mean)
kl_loss = - 0.5 * K.sum(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis=-1)
vae_loss = K.mean(xent_loss + kl_loss)
```

```python
vae.add_loss(vae_loss)
```

```python
from tensorflow.keras.optimizers import RMSprop
vae.compile(optimizer=RMSprop(lr=0.1))
vae.summary()
```

```python
vae.fit(X_train,
        shuffle=True,
        epochs=epochs,
        batch_size=256,
        validation_data=(X_test, None))
```
