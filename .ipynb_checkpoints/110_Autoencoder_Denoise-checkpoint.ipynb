# 노이즈 제거 오토인코더 (Denoising Autoencoder) 

deep autoencoder 를 훈련시키고, 얼굴 및 유사 이미지 탐색에 적용한다. 

[lfw dataset](http://vis-www.cs.umass.edu/lfw/).(Labeled Faces in the World) 의 사람 얼굴 이미지 이용.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Flatten, Reshape, Input, Conv2D, MaxPooling2D, UpSampling2D
import numpy as np
from sklearn.model_selection import train_test_split
from lfw_dataset import load_lfw_dataset
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

# Load dataset
Dataset was downloaded for you. Relevant links (just in case):

```python
# load images
X, attr = load_lfw_dataset(use_raw=True, dimx=32, dimy=32)
IMG_SHAPE = X.shape[1:]

# center images
X = X.astype('float32') / 255.

# split
X_train, X_test = train_test_split(X, test_size=0.1, random_state=42)
```

```python
plt.title('sample images')

for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(X[i])

# try to free memory
del X
import gc
gc.collect()
```

# convolutional autoencoder

Layer 들을 stack 하여 deep convolutional autoencoder 작성.

## Encoder


**encoder** 부분은 conv layer 와 pooling layer 를 stack. `activation='elu'` 를 사용한다.

kernel size (3, 3), `padding='same'` 으로 (conv, pool) 4 회 반복. 

## Decoder

**decoder** 는 **encoder** 의 반대로 layer 배치하고 UpSampling2D 사용.

```python
inp = Input(shape=(IMG_SHAPE))

x = Conv2D(32, (3,3), activation='relu', padding='same')(inp)
x = MaxPooling2D((2,2))(x)
x = Conv2D(64, (3,3), activation='relu', padding='same')(x)
x = MaxPooling2D((2,2))(x)
x = Conv2D(128, (3,3), activation='relu', padding='same')(x)
x = MaxPooling2D((2,2))(x)
x = Conv2D(256, (3,3), activation='relu', padding='same')(x)
encoded = MaxPooling2D((2,2))(x)

x = Conv2D(256, (3,3), activation='relu', padding='same')(encoded)
x = UpSampling2D((2,2))(x)
x = Conv2D(128, (3,3), activation='relu', padding='same')(x)
x = UpSampling2D((2,2))(x)
x = Conv2D(64, (3,3), activation='relu', padding='same')(x)
x = UpSampling2D((2,2))(x)
x = Conv2D(32, (3,3), activation='relu', padding='same')(x)
x = UpSampling2D((2,2))(x)
decoded = Conv2D(3, (3,3), activation='sigmoid', padding='same')(x)

autoencoder = Model(inp, decoded)
autoencoder.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
autoencoder.fit(x=X_train, y=X_train, epochs=50, validation_data=[X_test, X_test], verbose=2)
```

```python
reconstruction_mse = autoencoder.evaluate(X_test, X_test, verbose=0)
print("Convolutional autoencoder MSE:", reconstruction_mse)
for i in range(5):
    plt.subplot(1,3,1)
    plt.title("Original")
    plt.imshow((X_test[i]))
    
    plt.subplot(1,3,2)
    plt.title("Reconstructed")
    out = autoencoder.predict(np.expand_dims(X_test[i], axis=0))
    plt.imshow(np.squeeze(out, axis=0))
    plt.show()
```

# Denoising Autoencoder


autoencoder 를 이용하여 noise 를 제거한다.
<img src="denoising.jpg" style="width:40%">

model architecture 는 유지하지만 train 방식을 바꾼다. 각 epoch 의 시작 전에 입력 image 에 ranom 한 noise 를 섞는다.

noise 생성 방법: gaussian white noise 첨가

```python
def apply_gaussian_noise(X, sigma=0.1):
    """
    표준정규분포와 표준편차로 noise 추가
    :param X: image tensor of shape [batch,height,width,3]
    Returns X + noise.
    """
    noise = np.random.normal(loc=0.0, scale=sigma, size=X.shape)
    return X + noise
```

```python
# test different noise scales
plt.subplot(1,4,1)
plt.imshow(X_train[0])
plt.subplot(1,4,2)
plt.imshow(apply_gaussian_noise(X_train[:1],sigma=0.01)[0])
plt.subplot(1,4,3)
plt.imshow(apply_gaussian_noise(X_train[:1],sigma=0.1)[0])
plt.subplot(1,4,4)
plt.imshow(apply_gaussian_noise(X_train[:1],sigma=0.5)[0])
```

```python
for i in range(25):
    print("Epoch %i/25, Generating corrupted samples..."%(i+1))
    X_train_noise = apply_gaussian_noise(X_train)
    X_test_noise = apply_gaussian_noise(X_test)
    
    # we continue to train our model with new noise-augmented data
    autoencoder.fit(x=X_train_noise, y=X_train, epochs=1,  validation_data=[X_test_noise, X_test], verbose=2)
```

```python
X_test_noise = apply_gaussian_noise(X_test)
denoising_mse = autoencoder.evaluate(X_test_noise, X_test, verbose=0)
print("Denoising MSE:", denoising_mse)

for i in range(5):
    plt.subplot(1,3,1)
    plt.title("Original")
    plt.imshow((X_test_noise[i]))
    
    plt.subplot(1,3,2)
    plt.title("Reconstructed")
    out = autoencoder.predict(np.expand_dims(X_test_noise[i], axis=0))
    plt.imshow(np.squeeze(out, axis=0))
    plt.show()
```

# Cheap image morphing

- Image morphing : 컴퓨터 그래픽에서 어떤 모습을 전혀 다른 모습으로 변화시키는 것을 천천히 보여주는 기법

decoder 를 이용하여 새로운 image 생성하는 image code 들의 linear combination 을 취할 수 있다.

```python
np.linspace(0, 1,num=7)
```

```python
for _ in range(5):
    
    code1, code2 = X_test[np.random.randint(0, len(X_test), size=2)]   # image 2 개를 random select

    plt.figure(figsize=[10,4])
    
    for i, a in enumerate(np.linspace(0, 1,num=7)):

        output_code = code1*(1-a) + code2*(a)                                        # image 2 개를 mix
        output_image = autoencoder.predict(output_code[None])[0]           # mix 한 image 를 decode

        plt.subplot(1,7,i+1)                                                           
        plt.imshow(output_image)
        plt.title("a=%.2f"%a)
        
    plt.show()
```
